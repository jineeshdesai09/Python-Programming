```python
import math
x1= int(input("Enter x cordinate of point1: "))
y1= int(input("Enter y cordinate of point1: "))
x2= int(input("Enter x cordinate of point2: "))
y2= int(input("Enter y cordinate of point2: "))
dis= math.sqrt((x1-x2)**2 + (y1-y2)**2)
print(f"Distance between points is: {dis}")
```
